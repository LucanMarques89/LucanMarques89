```python
origem01 = '/content/drive/My Drive/Data/PRF/2024/'
origem02 = '/content/drive/My Drive/Data/PRF/2025/'
origem03 = '/content/drive/My Drive/Data/PRF/2026/'
```

```python
acidentes_todas_causas_tipos24 = origem01 + 'acidentes2024_todas_causas_tipos.csv'
acidentes_todas_causas_tipos25 = origem02 + 'acidentes2025_todas_causas_tipos.csv'
acidentes_todas_causas_tipos26 = origem03 + 'acidentes2026_todas_causas_tipos.csv'

acidentes24 = origem01 + 'acidentes2024.csv'
acidentes25 = origem02 + 'acidentes2025.csv'
acidentes26 = origem03 + 'acidentes2026.csv'

datatran24 = origem01 + 'datatran2024.csv'
datatran25 = origem02 + 'datatran2025.csv'
datatran26 = origem03 + 'datatran2026.csv'
```

```python
acidentes_todas_causas_tipos24 = pd.read_csv(acidentes_todas_causas_tipos24, low_memory=False, sep=';', encoding='latin1')
acidentes_todas_causas_tipos25 = pd.read_csv(acidentes_todas_causas_tipos25, low_memory=False, sep=';', encoding='latin1')
acidentes_todas_causas_tipos26 = pd.read_csv(acidentes_todas_causas_tipos26, low_memory=False, sep=';', encoding='latin1')

acidentes24 = pd.read_csv(acidentes24, low_memory=False, sep=';', encoding='latin1')
acidentes25 = pd.read_csv(acidentes25, low_memory=False, sep=';', encoding='latin1')
acidentes26 = pd.read_csv(acidentes26, low_memory=False, sep=';', encoding='latin1')

datatran24 = pd.read_csv(datatran24, low_memory=False, sep=';', encoding='latin1')
datatran25 = pd.read_csv(datatran25, low_memory=False, sep=';', encoding='latin1')
datatran26 = pd.read_csv(datatran26, low_memory=False, sep=';', encoding='latin1')
```
